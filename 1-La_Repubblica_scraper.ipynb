```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time
```

```python
data = []
fromdate = '1984-01-02'
page = 1

while True:
    url = f'https://ricerca.repubblica.it/ricerca/repubblica?query=berlusconi&fromdate={fromdate}&todate=2023-07-01&sortby=adate&author=&mode=any&page={page}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    articles = soup.find_all('article')
    print(f"Searching {page} for {fromdate}")

    if len(articles) == 0:
        # No articles found on the page, save the results and break the loop
        break
    
    for article in articles:
        title = article.find('a').text.strip()
        time_tag = article.find('time')
        if time_tag:
            date = time_tag['datetime']
        else:
            date = None
        url = article.find('a')['href']
        
        if 'berlusconi' in title.lower() or 'silvio' in title.lower():
            data.append({'Title': title, 'Date': date, 'URL': url})
        
        time.sleep(1)
    
    page += 1
    
    if page % 51 == 0:
        fromdate = date
        page = 1


df = pd.DataFrame(data)
df
```
